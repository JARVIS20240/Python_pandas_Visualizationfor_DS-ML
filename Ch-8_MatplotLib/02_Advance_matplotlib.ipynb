```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv("https://drive.google.com/uc?id=13cR6Fc1QUx7P23JTsyh28dxlHUD9Qeti")
```

```python
df.head()
```

```python
# Label Encoding :
df['Species'] = df['Species'].replace({"Iris-setosa":0, "Iris-versicolor" :1, "Iris-virginica":2})
df['Species'].value_counts()
```

```python
# SepalLengthCm & PetalLengthCm correlation :

plt.figure(figsize=(8,5))
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], c=df['Species'], cmap="viridis", alpha=1, label = [{0 :"Iris-setosa",1 : "Iris-versicolor",2 : "Iris-virginica"}])
# c= for seperate data here iris flower Species;
# cmap = for changing color them
# alpha = for opacity
plt.title("SepalLengthCm & PetalLengthCm correlation", fontsize="16")
plt.xlabel("SapelLength Cm")
plt.ylabel("PetalLength Cm")
plt.legend()
plt.colorbar() #show color bar to identify the data
```

```python
batter = pd.read_csv("https://drive.google.com/uc?id=1bZT5Hd48jJKVp3Cf8PhNS__-E5IeY9bD")
```

```python
batter.head()
```

```python
sample_df = batter.head(100).sample(25, random_state=5, ignore_index=True)
sample_df.head()
```

```python
x = [1,2,3]
y = [4,5,6]

plt.scatter(x,y)
plt.xticks([0,1,2,3,4,5])
plt.yticks([1,2,3,4,5,6])
plt.text(1, 4, "1st data")
plt.text(2, 5, "2st data")
plt.text(3, 6, "3st data")
plt.show()
```

```python
plt.figure(figsize=(18,10))
plt.title("Average vs Strick_rate Based Total Runs")
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])
plt.xlabel("Average")       
# plt.text(data_x, data_y, labale=string) = for give data to lable
plt.ylabel("Strick Rate")

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```
