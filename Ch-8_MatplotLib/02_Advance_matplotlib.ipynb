```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("fivethirtyeight")
```

```python
df = pd.read_csv("https://drive.google.com/uc?id=13cR6Fc1QUx7P23JTsyh28dxlHUD9Qeti")
```

```python
df.head()
```

```python
# Label Encoding :
df['Species'] = df['Species'].replace({"Iris-setosa":0, "Iris-versicolor" :1, "Iris-virginica":2})
df['Species'].value_counts()
```

### Data Seperate with Diff. Color code:

```python
# SepalLengthCm & PetalLengthCm correlation :

plt.figure(figsize=(8,5))
plt.scatter(df['SepalLengthCm'], df['PetalLengthCm'], c=df['Species'], cmap="viridis", alpha=1, label = [{0 :"Iris-setosa",1 : "Iris-versicolor",2 : "Iris-virginica"}])
# c= for seperate data here iris flower Species;
# cmap = for changing color them
# alpha = for opacity
plt.title("SepalLengthCm & PetalLengthCm correlation", fontsize="16")
plt.xlabel("SapelLength Cm")
plt.ylabel("PetalLength Cm")
plt.legend()
plt.colorbar() #show color bar to identify the data
```

## Change Data Dot's Size:

```python
plt.figure(figsize=(18,10))
plt.title("Average vs Strick_rate Based Total Runs")
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs']) #s= reprecent size of scatter Dots
plt.xlabel("Average")       
# plt.text(data_x, data_y, labale=string) = for give data to lable
plt.ylabel("Strick Rate")
# plt.grid(True, which="both")

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

## 2x2 Graph

##### 1st Way:

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(15,11))
ax[0,0].scatter(batter['avg'], batter['strike_rate'])
ax[0,0].set_title("Avg vs Strick Rate")
ax[0,0].set_xlabel("Average (0,0)")
ax[0,0].set_ylabel("Strick_Rate")

ax[0,1].scatter(batter['avg'], batter['runs'])
ax[0,1].set_title("Avg vs Runs")
ax[0,1].set_xlabel("Average (0,1)")
ax[0,1].set_ylabel("Runs")

ax[1,0].hist(batter['avg'])
ax[1,0].set_title("Avg Count")
ax[1,0].set_xlabel("Average (1,0)")
ax[1,0].set_ylabel("Count")

ax[1,1].hist(batter['runs'])
ax[1,1].set_title("Runs")
ax[1,1].set_xlabel("Runs (1,1)")
ax[1,1].set_ylabel("Count")
```
