```python
batter = pd.read_csv("https://drive.google.com/uc?id=1kJ1v2GDxa5uADs3uproeMp588RtRggJ_")
batter
```

```python
df = batter.sort_values(by=['avg', 'strike_rate'], ascending=False, ignore_index=True).head(50)
```

```python
plt.scatter(df['avg'], df['strike_rate'], marker='*')
plt.title("Average vs Starick Rate Analysis")
plt.axhline(120, color='g') # Batsman abow 120 Runs
plt.axvline(45, color='g') # Batsman Average abow 45

plt.xlabel("Average")
plt.ylabel("Straick Rate")
plt.show()
```

## Bar Plot:

    - Bivariate Analysis
    - Numerical vs Catagorical
    - Use: Aggrigate analysis for Groups

```python
# Simple Bar plot:
children = [10,20,40,30,10]
colors = ["red", "blue", "green", "yellow", 'pink']

plt.xlabel("Colors")
plt.ylabel("Children")
plt.bar(colors, children, color="black")
```

```python
# Horizontal Bar plot:
plt.xlabel("Colors")
plt.ylabel("Children")
plt.barh(colors, children, color="black")
```

```python
batsman_record = pd.read_csv("D:/Code/Python/PY_PANDAS/Ch-8_MatplotLib/CSVs/batsman_season_record.csv")
```

```python
batsman_record
```

```python
plt.title("Batsman Runs vs Season")
plt.xlabel("Batsman")
plt.ylabel("Seasons")
plt.bar(np.arange(batsman_record.shape[0]) - 0.3, batsman_record['2015'], width=0.3, label = "2015")
plt.bar(np.arange(batsman_record.shape[0]), batsman_record['2016'], width=0.3, label = "2016")
plt.bar(np.arange(batsman_record.shape[0]) + 0.3, batsman_record['2017'], width=0.3, label = "2017")
plt.legend()
plt.xticks(np.arange(batsman_record.shape[0]), batsman_record['batsman'])
plt.show()
```

```python
children = [10,20,40,30,10]
colors = ["red red red red", "blue blue blue blue", "green green green green", "yellow yellow yellow", 'pink pink pink pink']
c = ['red', 'blue', 'green', 'yellow', 'pink']
plt.xlabel("Colors")
plt.ylabel("Children")
plt.bar(colors, children, color= c)
plt.xticks(rotation="vertical")
plt.show()
```

```python
# Stack Bar plot:

plt.title("Batsman Runs vs Season")
plt.xlabel("Batsman")
plt.ylabel("Seasons")
plt.bar(batsman_record['batsman'], batsman_record['2017'], label="2017")
plt.bar(batsman_record['batsman'], batsman_record['2016'], bottom=batsman_record['2017'], label="2016")
plt.bar(batsman_record['batsman'], batsman_record['2015'], bottom=(batsman_record['2016'] + batsman_record['2017']), label="2015")
plt.legend()
plt.xticks(np.arange(batsman_record.shape[0]), batsman_record['batsman'])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Python', 'c', 'java', 'c++']
y = [85, 70, 60, 82]

plt.xlabel('Language', fontsize='10')
plt.ylabel('No.', fontsize='10')
plt.title('Languge demo', loc='left', fontsize='20')
plt.grid(False)

c = ['y', 'r', 'm', 'g']
plt.bar(x, y, width=0.4, color=c,edgecolor='r',linewidth=1, linestyle='--', alpha=0.5) 
# color = 'r' --> all colums are red , 
# Aligh bydefault 'cetner' 
# for blur use = alpha and alpha value is only possible between 0 to 1 

plt.show()
```

```python
x = ['Python', 'c', 'java', 'c++']
y = [85, 70, 60, 82]
c = ['y', 'c', 'pink', 'g']

plt.title('Language With Marks', fontsize= 15, loc="right")
plt.xlabel('Language', fontsize=12, loc="right")
plt.ylabel('Marks', fontsize=12, loc="top")
plt.bar(x, y, color=c,width=0.5, edgecolor='b', linewidth=1)
```

```python
import matplotlib.pyplot as plt

x = ['Python', 'c', 'java', 'c++']
y = [85, 70, 60, 82]
z = [40, 30, 20, 50]

plt.xlabel('Language', fontsize='10')
plt.ylabel('No.', fontsize='10')
plt.title('Languge demo', loc='left', fontsize='20')
plt.grid(False)

# Two Line in same column
plt.bar(x, y, width=0.4, color='g', label="laksh") 
plt.bar(x, z, width=0.4, color='r', label="laksh1")

plt.legend()
plt.show()
```
